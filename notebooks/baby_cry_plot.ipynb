```python
import json, time
import matplotlib.pyplot as plt
from IPython.display import clear_output

RESULTS_FILE = "D:\app project\currency-converter\results.json"
categories = ["Babbling", "Discomfort", "Hungry", "Cooing", "Belly Pain", "Tired"]

while True:
    try:
        with open(RESULTS_FILE, "r") as f:
            data = json.load(f)
    except (FileNotFoundError, json.JSONDecodeError):
        print("‚ö†Ô∏è Waiting for valid results.json ...")
        time.sleep(3)
        continue

    if not data:
        print("üì≠ No predictions yet. Waiting for updates ...")
        time.sleep(3)
        continue

    recent_data = data[-20:]
    times = [d["datetime"] for d in recent_data]
    y_values = [categories.index(d["result"]) if d["result"] in categories else None for d in recent_data]

    clear_output(wait=True)
    plt.figure(figsize=(10, 5))
    plt.plot(times, y_values, marker="o", color="dodgerblue", linewidth=2)
    plt.yticks(range(len(categories)), categories)
    plt.xticks(rotation=45, ha="right")
    plt.title("üçº Live Baby Cry Prediction Graph")
    plt.xlabel("Time")
    plt.ylabel("Prediction")
    plt.tight_layout()
    plt.show()

    time.sleep(5)
```
