```python
import json, time, matplotlib.pyplot as plt
from IPython.display import clear_output

while True:
    try:
        with open("results.json") as f:
            results = json.load(f)

        timestamps = [r["timestamp"] for r in results[:10]][::-1]
        preds = [r["prediction"] for r in results[:10]][::-1]

        clear_output(wait=True)
        plt.figure(figsize=(8,4))
        plt.plot(timestamps, range(len(preds)), 'o-', color='royalblue')
        plt.title("Recent Predictions (Live Update)")
        plt.xlabel("Timestamp")
        plt.ylabel("Prediction Index")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    except FileNotFoundError:
        print("⚠️ results.json not found. Please make sure your Flask app has generated it.")

    time.sleep(5)
```
